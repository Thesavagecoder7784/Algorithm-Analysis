# Binary Search

```python
# Binary Search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
# Binary Search Example
sorted_array = [1, 3, 5, 7, 9, 11, 13, 15]
target = 7
print("Binary Search:", binary_search(sorted_array, target))
```

# Breadth-First Search

```python
# Breadth-First Search (BFS)
from collections import deque
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph[node])

    return result
```

```python
# Graph for BFS and DFS
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B'],
    'F': ['C']
}

# BFS Example
print("BFS:", bfs(graph, 'A'))
```

# Depth-First Search

```python
# Depth-First Search (DFS)
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    result = [start]

    for neighbor in graph[start]:
        if neighbor not in visited:
            result.extend(dfs(graph, neighbor, visited))

    return result
```

```python
# DFS Example
print("DFS:", dfs(graph, 'A'))
```
